```python
import seaborn as sns
import plotly.graph_objects as go
import pandas as pd

df = sns.load_dataset("flights")
df
```

**Transformación de datos**

```python
df_transformado = df.pivot(index="year", columns="month", values="passengers")
df_transformado
```

**Mapa de calor**

```python
fig = go.Figure(data=go.Heatmap(z=df_transformado.values, x=df_transformado.columns, y=df_transformado.index, colorscale="Portland"))

fig.update_layout(
    title="Pasajeros por mes y año",
    xaxis_title="Mes",
    yaxis_title="Año",
)

fig.show()
```
