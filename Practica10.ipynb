```python
import seaborn as sns
import plotly.express as px

df = sns.load_dataset("tips")
df
```

**Histograma de propinas**

```python
fig = px.histogram(df, x="tip", color="day", nbins=20)

fig.update_layout(
    title="Histograma de propinas",
    xaxis_title="Propina ($)",
    yaxis_title="Frecuencia",
)

fig.show()
```
